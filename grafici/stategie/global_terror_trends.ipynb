```python
import pandas as pd
import matplotlib.pyplot as plt

# Carica i dati
df = pd.read_csv('../../data/globalterrorismdb.csv', encoding='latin1', low_memory=False)

# numero di attacchi per anno
attacks_per_year = df['iyear'].value_counts().sort_index()

plt.figure(figsize=(12,6))
attacks_per_year.plot(kind='line', color='teal', marker='o')
plt.title('Numero di attacchi terroristici per anno')
plt.xlabel('Anno')
plt.ylabel('Numero di attacchi')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# numero di vittime stimate per anno

victims_per_year = df.groupby('iyear')['nkill'].sum()

plt.figure(figsize=(12,6))
victims_per_year.plot(kind='line', color='crimson', marker='o')
plt.title('Numero di vittime (uccisi) per anno')
plt.xlabel('Anno')
plt.ylabel('Numero di vittime')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Serie gi calcolate
attacks_per_year = df['iyear'].value_counts().sort_index()
victims_per_year = df.groupby('iyear')['nkill'].sum()

# Creazione della figura e del primo asse
fig, ax1 = plt.subplots(figsize=(12,6))

# Primo grafico: numero di attacchi
color1 = 'teal'
ax1.set_xlabel('Anno')
ax1.set_ylabel('Numero di attacchi', color=color1)
ax1.plot(attacks_per_year.index, attacks_per_year.values, color=color1, marker='o', label='Attacchi')
ax1.tick_params(axis='y', labelcolor=color1)
ax1.grid(True, linestyle='--', alpha=0.5)

# Secondo asse condividendo l'asse x
ax2 = ax1.twinx()
color2 = 'crimson'
ax2.set_ylabel('Numero di vittime', color=color2)
ax2.plot(victims_per_year.index, victims_per_year.values, color=color2, marker='o', label='Vittime')
ax2.tick_params(axis='y', labelcolor=color2)

# Titolo
plt.title('Attacchi terroristici e vittime per anno')

# Legenda combinata
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper left')

plt.tight_layout()
plt.show()
```
